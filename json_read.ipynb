```python
import json
import cv2
from google.colab.patches import cv2_imshow 

with open("/content/dataset/label/101_00c3c3ae-4d99-4416-9b76-fdccf72f88ba.json", "r", encoding="utf8") as f:
   contents = f.read()
   json_data = json.loads(contents)

bbox=[]

file_name = json_data["images"][0]["file_name"] # image file_name
file_nm = file_name.split('.')
width = json_data["images"][0]["width"]   # image width
height = json_data["images"][0]["height"] # image height
ratio = 800.0 / width
dimension = (800, int(height * ratio))

image = cv2.imread('/content/dataset/img/' + file_nm[0] + '.jpg')

for i in range(len(json_data["annotations"])):
   bbox = json_data["annotations"][i]["bbox"]
   image = cv2.rectangle(image, (int(bbox[0]), int(bbox[1])), (int(bbox[0]+bbox[2]), int(bbox[1]+bbox[3])), (0, 255, 0), 2)

resized = cv2.resize(image, dimension, interpolation = cv2.INTER_AREA)
cv2_imshow(resized)
```
